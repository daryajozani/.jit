```python
from google.colab.patches import cv2_imshow
import cv2

# Download sample video
!curl -o daryavideo.mp4 /content/daryavideo.mp4

cap = cv2.VideoCapture('daryavideo.mp4')
while cap.isOpened():
    ret, image = cap.read()
    #gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    if not ret:
      break

    cv2_imshow(image) # Note cv2_imshow, not cv2.imshow
     #cv2_imshow(image)
    cv2.waitKey(1) & 0xff

cv2.destroyAllWindows()
cap.release()
```

```python
import cv2
from google.colab.patches import cv2_imshow
!curl -o dary1.jpg /content/darya1.jpg
!curl -o un.png /content/ct.jpg
img1=cv2.imread('dary1.jpg')
img2=cv2.imread('ct.jpg')
rows,cols,channels =img2.shape
#print("rows:",rows)
#print("cols:",cols)
#print("channels:",channels)
#print("________________________")
rol=img2[0:rows,0:cols]
#print(rol)
img2gray = cv2.cvtColor(img2,cv2.COLOR_BGR2GRAY)
ret,mask=cv2.threshold(img2gray,50,255,cv2.THRESH_BINARY_INV)
ret2,mask2=cv2.threshold(img2gray,50,255,cv2.THRESH_BINARY)
cv2_imshow(img2gray)
cv2_imshow(mask)
cv2_imshow(mask2)
cv2.waitKey(0)
```
